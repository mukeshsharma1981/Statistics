**Goodness of fit HT**

***In this type of HT, we aim to fit a set of candidate values of probability to the multiple classes of a multinomial random variable. ***

***Let p1, ..., pk be respectively the probability of class one to k. . ***

1. population proportion as p1 and p2
2. sample proportion as p̂1 and p̂2
3. sample size as n1 and n2

**Two Sided Hypothesis**
1. Step 1: State the hypotheses 

**H0: H0: p1 = p01,  p2 = p02, ..., pk = p0k **  

**H1:pj ≠ p0j, j = 1, ..., k, at least for one j**

2. Step 2: Set significance level

3. Step 3: Compute the test.value

** image.png **


4. Step 4: Find the c.value

**c.value = qchisq(1 – α, k – 1)**

5. Step 5: Specify the decision rule

if test.value ≥ c.value therefore H0 is rejected

6. Step 6: Make a decision and conclusion

If the condition in step 5 is satisfied, then the candidate values (p01, ..., p0k) are not a suitable choice for the set of parameters in the multinomial model at the significance level α.

**Example**

Use the boston_house_prices dataset from the folder of working datasets on the Moodle page.

Test whether each class of the variable RAD has the same probability at the 10% level of significance.   


```python
%load_ext rpy2.ipython
```

```python
%%R

data=read.csv('/content/boston_house_prices.csv')
head(data)

X=data$RAD

alpha = 0.10

O=table(X)

p=rep(1/9,9)

n=length(X)

E=n*p

test.value = sum((O-E)^2/E)

c.value = qchisq(1-alpha, (n-1))

c(test.value, c.value)

## since test.value < c.value so H0 is accepted which means 
## each class of the variable RAD has the same probability at the 10% level of significance.
```
