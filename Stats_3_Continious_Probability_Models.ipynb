# Continious Probability Model - Uniform Model

1. Data Simulation - runif(s,a,b)
*  s=sample size
*  a=the minimum value of the random variable
*  b=the maximum value of the random variable



2. Probability density Function (**pdf**) - dunif(j,a,b)
*  j=is the possible value of the ramdon variable
*  a=the minimum value of the random variable
*  b=the maximum value of the random variable



3. Cumulative Distribution Function (**cdf**) - punif(j,a,b)
*  j=is the possible value of the ramdon variable
*  a=the minimum value of the random variable
*  b=the maximum value of the random variable


## Example
Let random variable X be distributed with the Uniform model U(a=4,b=6.5). 

a) Simulate 100 samples from the above model.

b) Find P(X<5.3).

```python
%load_ext rpy2.ipython
```

```python
%%R
s=100;a=4;b=6.5
# a
print(runif(s,a,b))

# b
print(punif(5.3,a,b))
```

# Continious Probability Model - Normal Model

1. Data Simulation - rnorm(s, Âµ, Ïƒ)
*  s=sample size
*  Âµ=the parameter mean
*  Ïƒ=the parameter standard deviation



2. Probability density Function (**pdf**) - dnorm(j, Âµ, Ïƒ)
*  j=is the possible value of the ramdon variable
*  Âµ=the parameter mean
*  Ïƒ=the parameter standard deviation



3. Cumulative Distribution Function (**cdf**) - CDF=P(X<=j) =  pnorm(j, Âµ, Ïƒ)
*  j=is the possible value of the ramdon variable
*  Âµ=the parameter mean
*  Ïƒ=the parameter standard deviation




## Example
Let X be distributed with N(Âµ=1, Ïƒ=0.5).

a) Simulate 20 samples from this model.

b) Find P(0.7 < X < 1.2).

```python
%%R
s=20;mu=1; sd=0.5

# a
rnorm(s, mu, sd)

# b
pnorm(1.2,mu,sd)-pnorm(0.7,mu,sd)
```

## Example

For a certain type of computer, the length of time between charges of the battery is normally distributed with a mean of 50 hours and a standard deviation of 15 hours. 

a) Find the probability that the length of time is between 50 and 70 hours.

b) Write a code to plot the above normal pdf. 

```python
%%R
mu=50;sigma=15
pnorm(70,mu,sigma)-pnorm(50,mu,sigma)
```

```python
%%R

# b
X=seq(10,90,0.1)
plot(X,dnorm(X,mu,sigma), type='l')
```

# Continious Probability Model - Exponential Model

2. Probability density Function (**pdf**) - rexp(j, Î»)

3. Cumulative Distribution Function (**cdf**) - CDF=P(X<=j) = pexp(j,Î»)

## Example

```python
%%R
lam=2.4;s=10

# a
rexp(s,lam)
```

# Continious Probability Model - Gamma Model

1. Data Simulation - rgamma(s, a, Î»)
*  s=sample size
*  a=the parameter counter
*  Î» (Lambda)= the parameter rate 



2. Probability density Function (**pdf**) - dgamma(j, a, Î»)
*  j=is the possible value of the ramdon variable
*  a=the parameter counter
*  Î» (Lambda)= the parameter rate 



3. Cumulative Distribution Function (**cdf**) - CDF=P(X<=j) = pgamma(j, Î±, Î»)
*  j=is the possible value of the ramdon variable
*  a=the parameter counter
*  Î» (Lambda)= the parameter rate 


## Example
Let us assume a multiagent system constitutes eight components in a series configuration. Each component measures the local capacity (ci ) with the rate Î»i =0.5+i, i=1,...,8. 

a) If y=c1+c2+...+c8, what is the model of the systemâ€™s capacity?

b) Find p(y>1/5).

```python
%%R
alpha=8
lam=0
for (i in 1:8){
    lam=lam+0.5+i
} 
lam
```

```python
%%R
1-pgamma(1/5,alpha, lam)
```

# Continious Probability Model - Chi-Square model

1. Data Simulation - rchisq(s, k)
*  s=sample size
*  k=the parameter, degree of freedom 



2. Probability density Function (**pdf**) - dchisq(j,k)
*  j=is the possible value of the ramdon variable
*  k=the parameter, degree of freedom 



3. Cumulative Distribution Function (**cdf**) - CDF=P(X<=j) = pchisq(j,k)
*  j=is the possible value of the ramdon variable
*  k=the parameter, degree of freedom 


## Example

Let X be a Chi-Square random variable with k=3.

a) Generate 20 samples for X. 

b) Find p(0.2<X<2.4).

c) Plot the pdf of X. 

```python
%%R
k=3;s=20

# a
print(rchisq(s, k))

# b
print(pchisq(2.4,k)-pchisq(0.2,k))

```

```python
%%R
# c
x <- seq(0,20,100)
k=3

plot(x, dchisq(x,k), type='l', lyt=2, xlab='x value', ylab='y value', main='chi-square pdf for k=3')
```

# Continious Probability Model - Beta model

1. Data Simulation - rbeta(s,Î±,Î²)
*  s=sample size
*  Î±=the first shaping parameter
*  Î²=the second shaping parameter 


2. Probability density Function (**pdf**) - dbeta(j,Î±,Î²)
*  j=is the possible value of the ramdon variable
*  Î±=the first shaping parameter
*  Î²=the second shaping parameter 



3. Cumulative Distribution Function (**cdf**) - CDF=P(X<=j) =pbeta(j,Î±,Î²)
*  j=is the possible value of the ramdon variable
*  Î±=the first shaping parameter
*  Î²=the second shaping parameter 

## Example
In a MSN, four sensors are connected independently. Let p be the probability of the sensor's connection distributed with Beta(2,5). 

a) Generate 10 samples of p and compute the mean of simulated data.

b) Using the estimated value of p, find the probability that at least two sensors are connected in MSN.

```python
%%R
s=10;alpha=2;beta=5

# a
sim=rbeta(s,alpha,beta)
phat=mean(sim)

# b
a=1 - pbinom(1,4,phat)
a
```
